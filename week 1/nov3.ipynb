### Import the Modules

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

#### Create a Blank Image

```python
blank = np.zeros((100,200),dtype='uint8')
cv.imshow("Balnk Image",blank)
cv.waitKey(0)
```

### Color Convert

```python
bgr = cv.imread("nature.jpg")
rgb = cv.cvtColor(bgr,cv.COLOR_BGR2RGB)
gray = cv.cvtColor(bgr,cv.COLOR_BGR2GRAY)
hsv = cv.cvtColor(bgr,cv.COLOR_BGR2HSV)
lab = cv.cvtColor(bgr,cv.COLOR_BGR2LAB)
```

```python
images = [rgb, bgr, gray, hsv, lab]
titles = ['RGB', 'BGR', 'Grayscale', 'HSV', 'LAB']

plt.figure(figsize=(15, 10)) 

for i in range(5):
    plt.subplot(2, 3, i + 1)
    
    plt.title(titles[i], fontsize=12)
    plt.axis('off')

    if titles[i] == 'Grayscale':
        plt.imshow(images[i], cmap='gray')
    else:
        plt.imshow(images[i])


plt.show()
```
